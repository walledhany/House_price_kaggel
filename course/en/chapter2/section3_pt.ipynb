```python
from transformers import  AutoTokenizer
toknizer=AutoTokenizer.from_pretrained("bert-base-cased")
sequence = ("انا اسمة وليد من مصر","I hate this so m!")
token=toknizer.tokenize(sequence)
print(token)
```

```python
tokrn=toknizer(sequence)
print(tokrn)
```

```python
ids=toknizer.convert_tokens_to_ids(tokrn)
print(ids)
```
